```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/cost_revenue_clean (1).csv')
df
```

```python
df.describe()
```

```python
df.shape
```

```python
#X=df["production_budget_usd"].values (to take values just like in array)
#X
#y=df["worldwide_gross_usd"].values
#y
x=DataFrame(df,columns=['production_budget_usd'])
y=DataFrame(df,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title("film cost vs gross worldwide film revenue")
plt.xlabel("production cost $")
plt.ylabel("world gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.plot(x,reg.predict(x),color='red',linewidth=4)
plt.title("film cost vs gross worldwide film revenue")
plt.xlabel("production cost $")
plt.ylabel("world gross $")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
reg.score(x,y)
```
